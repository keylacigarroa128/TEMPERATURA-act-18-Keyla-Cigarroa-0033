## 1. Fórmulas de Conversión

Para movernos entre las escalas Celsius (centígrados) y Fahrenheit, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O de forma simplificada: $°F = (°C \times 1.8) + 32$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$
    *(O de forma simplificada: $°C = \frac{°F - 32}{1.8}$)*

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit
1.  **Identificar datos:** Tenemos $25°C$.
2.  **Sustituir en la fórmula:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
5.  **Resultado:** **77°F**.

### Ejemplo 2: Convertir 98.6°F a Celsius
1.  **Identificar datos:** Tenemos $98.6°F$ (temperatura corporal promedio).
2.  **Sustituir en la fórmula:** $°C = (98.6 - 32) / 1.8$.
3.  **Restar primero (paréntesis):** $98.6 - 32 = 66.6$.
4.  **Dividir:** $66.6 / 1.8 = 37$.
5.  **Resultado:** **37°C**.

## 3. Código en Python

Este programa utiliza funciones para segmentar la lógica, haciendo que el código sea limpio y fácil de mantener.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- Resultado: {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- Resultado: {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Controla el flujo principal del programa"""
    while True:
        print("\n--- CALCULADORA DE TEMPERATURA ---")
        print("1. Convertir Grados Celsius a Fahrenheit")
        print("2. Convertir Grados Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
```

### Explicación técnica del código:
* **`while True`:** Crea un bucle infinito que permite al usuario realizar múltiples conversiones sin tener que reiniciar el programa.
* **`float()`:** Usamos `float` en lugar de `int` para permitir decimales (muy comunes en temperatura).
* **`f-strings` y `:.2f`:** Se utiliza para imprimir los resultados de forma elegante, limitando a 2 los decimales mostrados.
* **`break`:** Rompe el ciclo inmediatamente cuando el usuario elige la opción 3

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- Resultado: {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- Resultado: {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Controla el flujo principal del programa"""
    while True:
        print("\n--- CALCULADORA DE TEMPERATURA ---")
        print("1. Convertir Grados Celsius a Fahrenheit")
        print("2. Convertir Grados Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Keyla Paola Cigarroa Palacios NL = 12")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
````
